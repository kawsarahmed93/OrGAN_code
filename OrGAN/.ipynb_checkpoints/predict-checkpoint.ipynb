# PSNR function

```python
def PSNR(original, pred):
    mse = np.mean((original - pred) ** 2)
    if mse == 0:
        return 100
    PIXEL_MAX = 1.0
    PSNR = 20 * math.log10(PIXEL_MAX / math.sqrt(mse))
    return PSNR

def plot_images(images, cols=3):
#    plt.clf()
#    plt.figure(figsize(14,8))
    rows = math.ceil(images.shape[0]/cols)
    fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(14,5*rows))

    idx = 0
    for row in ax:
        for col in row:
            if idx > images.shape[0]-1:
                break
            im = images[idx,:,:]
            col.imshow(im, cmap=plt.cm.bone)
            col.title.set_text(idx)
            idx += 1
    plt.show()
```

```python
# model.eval()
# critic.eval()
num_val_batches = len(test_loader)
avg_loss1 = 0
avg_loss2 = 0
avg_psnr = 0
bce = torch.nn.BCELoss()
mse = torch.nn.MSELoss()
preds = []
xray = []
count = 0
# iterate over the validation set
for batch, (images, masks) in enumerate(test_loader):
    #(image, mask_true) = batch
    # move images and labels to correct device and type
    images = images.to(device=device, dtype=torch.float32)
    lungs = masks.unsqueeze(1).to(device=device, dtype=torch.float32)
    
    #bone = masks[1]
#     true_masks = torch.stack((lungs,bone)).permute(1, 0, 2, 3)
#     true_masks = true_masks.to(device=device, dtype=torch.float32)

    with torch.no_grad():
        
        mask_pred, _ = model(images)
        
        true = critic(lungs)
#         true = (true > 0.5)*1.0
        pred = critic(mask_pred)
#         pred = (pred > 0.5)*1.0
#         target_true = torch.ones_like(true_map)
#         print(mask_pred[:][0].unsqueeze(1).shape)
#         pred_map = critic(mask_pred[:][0].unsqueeze(1)).view(-1)
#         target_pred = torch.zeros_like(pred_map)
        
#         target_true = (lungs > 0.05).float()
#         target_pred = torch.zeros_like(mask_pred)
        
#         loss1 = bce(true_map, target_true) 
#         loss2 = bce(pred_map, target_pred) 
        
        mask_pred = mask_pred.cpu().detach().numpy()
        true = true.cpu().detach().numpy()
        pred = pred.cpu().detach().numpy()
#         print(mask_pred.shape)
#         pred_map = pred_map.cpu().detach().numpy().squeeze(0)
#         pred_map = pred_map > 0.4
        lungs = lungs.cpu().detach().numpy().squeeze(0)
        psnr =  PSNR(lungs, mask_pred)
#         true_map = true_map.cpu().detach().numpy().squeeze(0)
#         true_map = true_map > 0.4
#         mask_pred[mask_pred<0.03]=0
        images = images.cpu().detach().numpy().squeeze(0)
        
#         avg_loss1 += loss1.item()
#         avg_loss2 += loss2.item()
        avg_psnr += psnr
        
#         pred.append([mask_pred,pred_map,lungs,true_map])
        preds.append([pred,mask_pred,lungs,true])
        xray.append(images)        
        count = count + 1
        
        if count == 50:
            break
        
# if num_val_batches == 0:
#     print('true: ' + str(avg_loss1)+' pred: ' + str(avg_loss2)+' psnr: ' + str(avg_psnr))
print(' psnr: ' + str(avg_psnr/count))
```
