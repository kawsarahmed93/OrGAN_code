# PSNR function

```python
def PSNR(original, pred):
    mse = np.mean((original - pred) ** 2)
    if mse == 0:
        return 100
    PIXEL_MAX = 1.0
    PSNR = 20 * math.log10(PIXEL_MAX / math.sqrt(mse))
    return PSNR
```

```python
transforms = A.Compose(
    [
        A.Resize(width=512, height=512, p=1.0)
    ] 
)
```

```python
BATCH_SIZE_TEST = 1
model_iter = '1'

test_xray_directory = 'data/Test/Xray'
test_xray_filenames = os.listdir(test_xray_directory)

dataset_test = TXDataset(test_xray_filenames,test_xray_directory,transform = transforms)
```

```python
test_loader = DataLoader(dataset_test, BATCH_SIZE_TEST, shuffle = True)
```

# Model Creation

```python
model = OrGAN()
```

# Inference on simulated Validation set

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model= nn.DataParallel(model,device_ids = [0])
model  = torch.load('model_weights/'+model_iter+'/best.ckpt')
model = model.to(device)
```

```python
num_val_batches = len(test_loader)
avg_psnr = 0
bce = torch.nn.BCELoss()
mse = torch.nn.MSELoss()
preds = []
xray = []
count = 0
# iterate over the validation set
for batch, (images, masks) in enumerate(test_loader):
    images = images.to(device=device, dtype=torch.float32)
    lungs = masks.unsqueeze(1).to(device=device, dtype=torch.float32)

    with torch.no_grad():
        
        mask_pred, _ = model(images)
        
        mask_pred = mask_pred.cpu().detach().numpy()
        lungs = lungs.cpu().detach().numpy().squeeze(0)
        
        psnr =  PSNR(lungs, mask_pred)
        images = images.cpu().detach().numpy().squeeze(0)
        avg_psnr += psnr
        preds.append([mask_pred,lungs])
        xray.append(images)        
        count = count + 1
        
print(' psnr: ' + str(avg_psnr/count))
```

```python
image = []
for x in range(len(preds)):
    mask = preds[x][0].squeeze(0).squeeze(0)
    lung = preds[x][1].squeeze(0)
    image.append(xray[x][0])
    image.append(mask)    
    image.append(lung)
```
